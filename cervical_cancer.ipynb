# Problem Statement

Cervical cancer is one of the most common and life-threatening cancers affecting women worldwide. Early detection is critical for effective treatment and improving survival rates. However, manual diagnosis based on multiple medical and lifestyle factors can be time-consuming and may not always identify high-risk patients efficiently.

The objective of this project is to develop a machine learning-based predictive system that can analyze patient medical history, demographic information, and clinical test results to predict the likelihood of cervical cancer, specifically the **Biopsy result** (positive or negative).

The system uses features such as age, sexual history, smoking habits, contraceptive use, medical conditions, and screening test results to train and evaluate machine learning models. The goal is to build an accurate and reliable prediction model that can assist healthcare professionals in early detection, risk assessment, and decision-making, ultimately helping to improve patient outcomes and reduce cervical cancer mortality.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x=train_df.drop('Biopsy',axis=1)
y=train_df['Biopsy']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Baseline model accuracy: {accuracy:.2f}')
```

**DecisionTree**

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f'Baseline model accuracy: {accuracy:.2f}')
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', LinearSVC(max_iter=5000))
])

pipe.fit(x_train, y_train)
pipe.score(x_test, y_test)
```

```python
from sklearn.model_selection import KFold
```

```python
x=train_df.drop('Biopsy',axis=1)
y=train_df['Biopsy']

kf=KFold(n_splits=5,shuffle=True,random_state=42)
model=LogisticRegression()

best_accuracy=0
best_fold=-1

best_x_train=None
best_y_train=None
best_x_test=None
best_y_test=None
best_y_pred=None

for fold,(train_index,test_index) in enumerate(kf.split(x),1):
  x_train, x_test = x.iloc[train_index], x.iloc[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

  model.fit(x_train,y_train)

  y_pred=model.predict(x_test)
  acc=accuracy_score(y_test,y_pred)
  print(f"fold{fold} Accuracy:{acc}")

  if acc >best_accuracy:
    best_accuracy=acc
    best_fold=fold
    best_x_train=x_train
    best_y_train=y_train
    best_x_test=x_test
    best_y_test=y_test
    best_y_pred=y_pred
print(f"\n best fold:{best_fold} with accuracy:{best_accuracy}")
```

```python
model = LogisticRegression()
model.fit(best_x_train,best_y_train)
model.score(best_x_test,best_y_test)
```

```python
model = DecisionTreeClassifier()
model.fit(best_x_train,best_y_train)
model.score(best_x_test,best_y_test)
```

```python
model = RandomForestClassifier()
model.fit(best_x_train,best_y_train)
model.score(best_x_test,best_y_test)
```

```python
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', LinearSVC(max_iter=5000))
])

pipe.fit(best_x_train,best_y_train)
pipe.score(best_x_test,best_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
train_df.info()
```

```python
model=Sequential()
model.add(Dense(128,input_shape=(7,),activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(best_x_train,best_y_train,epochs=20,batch_size=32)
```

```python
loss,accuracy=model.evaluate(best_x_test,best_y_test)
print(f"Model Accuracy:{accuracy * 100:.2f}%")
```
